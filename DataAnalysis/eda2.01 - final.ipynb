# Feature Importance

```python
# Prepare data for analysis
from sklearn.preprocessing import StandardScaler
from scipy.stats import mannwhitneyu, pointbiserialr
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import mutual_info_classif, f_classif
from sklearn.inspection import permutation_importance
from sklearn.model_selection import cross_val_score
```

```python
# Separate features and target
X = df.drop(columns=["authorized"])
y = df["authorized"]
```

```python
# Standardize features for analysis
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# F-STATISTIC
f_scores, f_pvalues = f_classif(X_scaled, y)
f_results = pd.DataFrame({
    'feature': X.columns,
    'f_score': f_scores,
    'p_value': f_pvalues,
    'significant': f_pvalues < 0.05
}).sort_values('f_score', ascending=False)
```

```python
f_results = f_results.reset_index(drop=True)
print("F-statistic Feature Selection:")
print(f_results.round(4))
```

```python
# STATISTICAL SIGNIFICANCE TESTING
statistical_results = []

for feature in X.columns:
    benign_data = df[df['authorized'] == True][feature]
    violation_data = df[df['authorized'] == False][feature]
    
    # Mann-Whitney U test (non-parametric)
    try:
        statistic, p_value = mannwhitneyu(benign_data, violation_data, alternative='two-sided')
        
        # Effect size (Cohen's d)
        pooled_std = np.sqrt(((len(benign_data)-1)*benign_data.var() + 
                             (len(violation_data)-1)*violation_data.var()) / 
                            (len(benign_data) + len(violation_data) - 2))
        
        if pooled_std > 0:
            cohens_d = (benign_data.mean() - violation_data.mean()) / pooled_std
        else:
            cohens_d = 0
            
        # Effect size interpretation
        effect_size = 'Large' if abs(cohens_d) >= 0.8 else 'Medium' if abs(cohens_d) >= 0.5 else 'Small'
        
        statistical_results.append({
            'feature': feature,
            'benign_mean': benign_data.mean(),
            'violation_mean': violation_data.mean(),
            'p_value': p_value,
            'cohens_d': abs(cohens_d),
            'effect_size': effect_size,
            'significant': p_value < 0.05
        })
    except Exception as e:
        print(f"Error processing {feature}: {e}")
```

```python
stats_df = pd.DataFrame(statistical_results).sort_values('cohens_d', ascending=False)
stats_df = stats_df.reset_index(drop=True)
print("Statistical Test Results (sorted by effect size):")
print(stats_df.round(4))
```

```python
# Feature correlation matrix
plt.figure(figsize=(12, 10))
corr_matrix = X.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Point-biserial correlation with target
correlations = []
for feature in X.columns:
    try:
        corr, p_val = pointbiserialr(df['authorized'].astype(int), df[feature])
        correlations.append({
            'feature': feature, 
            'correlation': abs(corr), 
            'correlation_raw': corr,
            'p_value': p_val,
            'significant': p_val < 0.05
        })
    except Exception as e:
        print(f"Error calculating correlation for {feature}: {e}")
```

```python
corr_df = pd.DataFrame(correlations).sort_values('correlation', ascending=False)
corr_df = corr_df.reset_index(drop=True)
print("Point-biserial correlations with target (sorted by absolute correlation):\n")
print(corr_df.round(4))
```

```python
# RANDOM FOREST FEATURE IMPORTANCE
rf_importances = []
for i in range(10):  # Multiple runs for stability
    rf = RandomForestClassifier(n_estimators=100, random_state=i)
    rf.fit(X_scaled, y)
    rf_importances.append(rf.feature_importances_)
```

```python
# Average importance across runs
avg_importance = np.mean(rf_importances, axis=0)
std_importance = np.std(rf_importances, axis=0)
```

```python
rf_results = pd.DataFrame({
    'feature': X.columns,
    'importance': avg_importance,
    'std': std_importance
}).sort_values('importance', ascending=False)
```

```python
rf_results = rf_results.reset_index(drop=True)
print("Random Forest Feature Importance:")
print(rf_results.round(4))
```

```python
# Visualize RF importance
plt.figure(figsize=(10, 6))
plt.barh(range(len(rf_results)), rf_results['importance'], 
         xerr=rf_results['std'], alpha=0.8)
plt.yticks(range(len(rf_results)), rf_results['feature'])
plt.xlabel('Feature Importance')
plt.title('Random Forest Feature Importance (with std dev)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import plot_tree

# Train a single Random Forest model (optional: increase visibility by using fewer trees)
rf_final = RandomForestClassifier(n_estimators=5, random_state=42)
rf_final.fit(X_scaled, y)

# Plot the first 3 trees
for i in range(3):
    plt.figure(figsize=(20, 10))
    plot_tree(
        rf_final.estimators_[i],
        feature_names=X.columns,
        class_names=["Unauthorized", "Authorized"],
        filled=True,
        rounded=True,
        max_depth=3  # limit depth for readability
    )
    plt.title(f"Decision Tree {i+1} from Random Forest")
    plt.show()
```

```python
# MUTUAL INFORMATION
mi_scores = mutual_info_classif(X_scaled, y, random_state=42)
mi_results = pd.DataFrame({
    'feature': X.columns,
    'mutual_info': mi_scores
}).sort_values('mutual_info', ascending=False)
```

```python
mi_results = mi_results.reset_index(drop=True)
print("Mutual Information Scores:")
print(mi_results.round(4))
```

```python
# PERMUTATION IMPORTANCE
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_scaled, y)
```

```python
perm_importance = permutation_importance(rf, X_scaled, y, n_repeats=10, random_state=42)
perm_results = pd.DataFrame({
    'feature': X.columns,
    'importance': perm_importance.importances_mean,
    'std': perm_importance.importances_std
}).sort_values('importance', ascending=False)
```

```python
perm_results = perm_results.reset_index(drop=True)
print("Permutation Importance:")
print(perm_results.round(4))
```

```python
# INTEGRATED FEATURE RANKING
# Combine all methods for comprehensive ranking
combined_results = pd.DataFrame({'feature': X.columns})
combined_results = combined_results.merge(stats_df[['feature', 'cohens_d']], on='feature')
combined_results = combined_results.merge(corr_df[['feature', 'correlation']], on='feature')
combined_results = combined_results.merge(rf_results[['feature', 'importance']], on='feature')
combined_results = combined_results.merge(mi_results[['feature', 'mutual_info']], on='feature')
combined_results = combined_results.merge(perm_results[['feature', 'importance']], on='feature', suffixes=('_rf', '_perm'))
```

```python
# Normalize all scores to 0-1 scale for comparison
from sklearn.preprocessing import MinMaxScaler
scaler_norm = MinMaxScaler()
score_columns = ['cohens_d', 'correlation', 'importance_rf', 'mutual_info', 'importance_perm']
combined_results[score_columns] = scaler_norm.fit_transform(combined_results[score_columns])
```

```python
# Create composite score
combined_results['composite_score'] = combined_results[score_columns].mean(axis=1)
combined_results = combined_results.sort_values('composite_score', ascending=False)
```

```python
combined_results = combined_results.reset_index(drop=True)
print("Final Integrated Feature Ranking:")
print(combined_results[['feature', 'composite_score'] + score_columns].round(4))
```

```python
# Visualize integrated ranking
plt.figure(figsize=(12, 8))
plt.barh(range(len(combined_results)), combined_results['composite_score'], alpha=0.8)
plt.yticks(range(len(combined_results)), combined_results['feature'])
plt.xlabel('Composite Importance Score')
plt.title('Integrated Feature Importance Ranking')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Summary insights
top_3_features = combined_results.head(3)['feature'].tolist()
print(f"Top 3 most discriminative features: {', '.join(top_3_features)}")
```

```python
highly_significant = stats_df[stats_df['significant'] == True]['feature'].tolist()
print(f"Statistically significant features ({len(highly_significant)}): {', '.join(highly_significant)}")
```

# Dimensionality Reduction with UMAP

```python
import umap.umap_ as umap
from sklearn.preprocessing import StandardScaler
```

```python
# Prepare features and target
X = df.drop(columns=["authorized"])
y = df["authorized"]
```

```python
X_encoded = X.astype(int).values 
```

```python
# Apply UMAP
umap_model = umap.UMAP(
    n_neighbors=20, # can be modified from 10-200
    min_dist=0.5, # can be modified from 0.001 to 0.99
    n_components=2,
    random_state=42
)

X_umap = umap_model.fit_transform(X_encoded)
```

```python
# Create DataFrame for plotting
umap_df = pd.DataFrame(X_umap, columns=["UMAP1", "UMAP2"])
umap_df["authorized"] = y

# Visualize UMAP projection
plt.figure(figsize=(10, 7))
sns.scatterplot(data=umap_df, x="UMAP1", y="UMAP2", hue="authorized", palette="Set2")
plt.title("UMAP Projection of All Features")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score, silhouette_score

# Evaluate clustering vs true labels
ari_score = adjusted_rand_score(y, umap_df["authorized"])  # Should be perfect since labels = labels
sil_score = silhouette_score(X_umap, y)  # Measures how well points of the same class group together

print(f"Adjusted Rand Index (ARI): {ari_score:.3f}")
print(f"Silhouette Score: {sil_score:.3f}")
```

# Clustering with k-means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Input: UMAP-transformed data (from previous UMAP step)
X_umap_only = umap_df[["UMAP1", "UMAP2"]].values
```

```python
# Range of K values to try
k_range = range(2, 15)

inertia_scores = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_umap_only)
    
    inertia_scores.append(kmeans.inertia_)
    
    if len(set(cluster_labels)) > 1:
        sil_score = silhouette_score(X_umap_only, cluster_labels)
        silhouette_scores.append(sil_score)
    else:
        silhouette_scores.append(-1)
```

```python
# Plot Inertia (Elbow Method)
plt.figure(figsize=(10, 4))
plt.plot(k_range, inertia_scores, marker="o")
plt.title("K-Means Inertia vs. Number of Clusters (K)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (WCSS)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot Silhouette Score
plt.figure(figsize=(10, 4))
plt.plot(k_range, silhouette_scores, marker="o", color="purple")
plt.title("Silhouette Score vs. Number of Clusters (K)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a summary DataFrame of silhouette scores
silhouette_table = pd.DataFrame({
    "K": list(k_range),
    "Silhouette Score": silhouette_scores
}).round(4)

# Display the table
print("\nSilhouette Scores by K:")
display(silhouette_table)
```

```python
# Visualize K-Means Clustering on UMAP Projection for a Specific K
k_visualize = 4 # can be modified based on the elbow

kmeans_vis = KMeans(n_clusters=k_visualize, random_state=42, n_init=10)
umap_df["kmeans_cluster"] = kmeans_vis.fit_predict(X_umap_only)

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=umap_df,
    x="UMAP1",
    y="UMAP2",
    hue="kmeans_cluster",
    palette="tab10",
    legend="full"
)
plt.title(f"K-Means Clustering on UMAP (K = {k_visualize})")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Clustering with DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
# UMAP result should already be stored in `umap_df`
X_umap_features = umap_df[["UMAP1", "UMAP2"]].values
y_true = umap_df["authorized"].values
```

```python
# Store results
eps_val = 4 # can be modified
min_sam = 5 # can be modified

db = DBSCAN(eps=eps_val, min_samples=min_sam)
clusters = db.fit_predict(X_umap_features)

umap_df["cluster"] = clusters
n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
n_noise = list(clusters).count(-1)

print(f"Clusters found: {n_clusters}")
print(f"Noise points: {n_noise}")

# ARI (how well clusters match "authorized" label)
ari = adjusted_rand_score(y_true, clusters)
print(f"Adjusted Rand Index (ARI): {ari:.4f}")

# Silhouette Score
if n_clusters > 1:
    sil_score = silhouette_score(X_umap_features, clusters)
    print(f"Silhouette Score: {sil_score:.4f}")
else:
    print("Silhouette Score not applicable (only 1 cluster).")

# Visualize
plt.figure(figsize=(10, 7))
sns.scatterplot(data=umap_df, x="UMAP1", y="UMAP2", hue="cluster", palette="tab20", legend="full")
plt.title(f"DBSCAN Clustering (eps={eps_val}, min_samples={min_sam})")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# General grouped plotting function
def plot_categorical_features_by_cluster(df, cluster_col, umap_cols, cluster_range, feature_map, title_group):
    for cluster_id in cluster_range:
        subset = df.copy()
        is_current_cluster = subset[cluster_col] == cluster_id

        plt.figure(figsize=(8, 6))

        # Plot background in gray
        sns.scatterplot(
            data=subset[~is_current_cluster],
            x=umap_cols[0],
            y=umap_cols[1],
            color="lightgray",
            alpha=0.4,
            label="Other Clusters"
        )

        # Overlay for each feature in this cluster
        for label, (col, color) in feature_map.items():
            mask = is_current_cluster & (subset[col] == 1)
            sns.scatterplot(
                data=subset[mask],
                x=umap_cols[0],
                y=umap_cols[1],
                color=color,
                label=label
            )

        plt.title(f"{title_group}\nCluster {cluster_id}")
        plt.xlabel(umap_cols[0])
        plt.ylabel(umap_cols[1])
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# General binary attribute plotter with label mapping
def plot_binary_feature_by_cluster(df, cluster_col, umap_cols, cluster_range, feature_col, value_labels, title_group, colors=("red", "blue")):
    for cluster_id in cluster_range:
        subset = df.copy()
        is_current_cluster = subset[cluster_col] == cluster_id

        plt.figure(figsize=(8, 6))

        # Plot background
        sns.scatterplot(
            data=subset[~is_current_cluster],
            x=umap_cols[0],
            y=umap_cols[1],
            color="lightgray",
            alpha=0.4,
            label="Other Clusters"
        )

        # Plot class 0 and 1 with mapped labels
        for val, label_text in value_labels.items():
            mask = is_current_cluster & (subset[feature_col] == val)
            sns.scatterplot(
                data=subset[mask],
                x=umap_cols[0],
                y=umap_cols[1],
                label=label_text,
                color=colors[val]
            )

        plt.title(f"{title_group}\nCluster {cluster_id}")
        plt.xlabel(umap_cols[0])
        plt.ylabel(umap_cols[1])
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# Cluster range
cluster_ids = sorted(merged_df["kmeans_cluster"].unique())

# Feature mappings
user_role_map = {
    "Admin Staff": ("is_user_administrative_staff", "red"),
    "Manager": ("is_user_manager", "orange"),
    "Director": ("is_user_director", "blue")
}

presence_map = {
    "AS Present": ("administrative_staff_presence", "red"),
    "Manager Present": ("manager_presence", "orange"),
    "Director Present": ("director_presence", "blue")
}

file_dest_map = {
    "To External Drive": ("is_file_went_to_external_drive", "red"),
    "To Other Staff": ("is_file_went_to_other_staff_drive", "orange"),
    "To Internal Drive": ("is_file_went_to_internal_drive", "blue")
}

file_action_map = {
    "Opened": ("is_file_opening", "red"),
    "Modified": ("is_file_modification", "orange"),
    "Deleted": ("is_file_deletion", "blue"),
    "Moved": ("is_file_move", "green"),
    "Copied": ("is_file_copy", "purple")
}

# Run grouped visualizations
plot_categorical_features_by_cluster(merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids, user_role_map, "User Role")
plot_categorical_features_by_cluster(merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids, presence_map, "User Presence")
plot_categorical_features_by_cluster(merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids, file_dest_map, "File Destination")
plot_categorical_features_by_cluster(merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids, file_action_map, "File Access Type")

# Run binary attribute visualizations
plot_binary_feature_by_cluster(
    merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids,
    "room_mismatch", {0: "Match", 1: "Mismatch"}, "Room Mismatch"
)

plot_binary_feature_by_cluster(
    merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids,
    "shift_status", {0: "Outside Shift", 1: "Within Shift"}, "Shift Status"
)

plot_binary_feature_by_cluster(
    merged_df, "kmeans_cluster", ["UMAP1", "UMAP2"], cluster_ids,
    "is_confidential", {0: "Non-confidential", 1: "Confidential"}, "File Confidentiality"
)
```
